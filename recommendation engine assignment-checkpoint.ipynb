### Use Euclideon Distance as a Metric for UBCF

```python
from sklearn.metrics import pairwise_distances
```

```python
euclidean_pivot_table = pd.pivot_table(data = book_data,values='Rating',index='UserID',columns='Title').fillna(0)
euclidean_pivot_table
```

```python
euclidean_pivot_table.index = book_data['UserID'].unique()
euclidean_pivot_table
```

```python
euclidean_pivot_table.values
```

```python
euclidean_ubcf = pairwise_distances(X=euclidean_pivot_table.values,metric='euclidean')
euclidean_ubcf
```

```python
Euclidean_ubcf_metric = pd.DataFrame(data = euclidean_ubcf )
Euclidean_ubcf_metric
```

```python
Euclidean_ubcf_metric.columns = book_data['UserID'].unique()
Euclidean_ubcf_metric.index = book_data['UserID'].unique()
Euclidean_ubcf_metric
```

### Use Cosine Distance as a Metric for UBCF

```python
cosine_pivot_table = pd.pivot_table(data = book_data,values='Rating',index='UserID',columns='Title').fillna(0)
cosine_pivot_table.index = book_data['UserID'].unique()
cosine_pivot_table
```

```python
cosine_pivot_table.values
```

```python
cosine_ubcf = 1 - pairwise_distances(X = cosine_pivot_table.values,metric='cosine')
cosine_ubcf
```

```python
cosine_ubcf_metric = pd.DataFrame(data = cosine_ubcf)
cosine_ubcf_metric.columns = book_data['UserID'].unique()
cosine_ubcf_metric.index = book_data['UserID'].unique()
cosine_ubcf_metric.round(2)
```

```python
np.fill_diagonal(a = cosine_ubcf_metric.to_numpy(), val = 0, wrap=False)
```
